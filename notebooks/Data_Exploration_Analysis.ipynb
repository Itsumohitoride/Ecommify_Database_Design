## 3. Detección de Outliers en DataFrames

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

dataframes = {
    'customers_df': customers_df,
    'geolocation_df': geolocation_df,
    'order_items_df': order_items_df,
    'order_payments_df': order_payments_df,
    'order_reviews_df': order_reviews_df,
    'orders_df': orders_df,
    'products_df': products_df,
    'sellers_df': sellers_df,
    'product_category_name_translation_df': product_category_name_translation_df
}

print("### Detección de Outliers utilizando el método IQR\n")

# Define columns to exclude from outlier detection
excluded_cols_from_outliers = [
    'order_item_id', 'payment_sequential',
    'customer_zip_code_prefix', 'seller_zip_code_prefix', 'geolocation_zip_code_prefix' # Zip codes are identifiers
]

for df_name, df in dataframes.items():
    print(f"\n--- Analizando DataFrame: {df_name} ---")
    numerical_cols = df.select_dtypes(include=['number']).columns

    # Filter out excluded columns
    cols_to_analyze = [col for col in numerical_cols if col not in excluded_cols_from_outliers]

    if not cols_to_analyze:
        print(f"No hay columnas numéricas relevantes para analizar outliers en {df_name} después de excluir IDs/secuenciales.")
        continue

    found_outliers = False
    for col in cols_to_analyze:
        outliers, lower_bound, upper_bound = detect_outliers_iqr(df, col)
        if not outliers.empty:
            found_outliers = True
            print(f"  Outliers encontrados en la columna '{col}':")
            print(f"    Rango normal (IQR): [{lower_bound:.2f}, {upper_bound:.2f}]")
            print(f"    Número de outliers: {len(outliers)}")
            display(outliers.head())

    if not found_outliers:
        print(f"  No se encontraron outliers significativos en las columnas numéricas de {df_name} (excluyendo IDs/secuenciales).")
```

## 3.1. Visualización de Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("### Visualización de Outliers mediante Box Plots\n")

# Define columns to exclude from outlier detection (should match the previous cell)
excluded_cols_from_outliers = [
    'order_item_id', 'payment_sequential',
    'customer_zip_code_prefix', 'seller_zip_code_prefix', 'geolocation_zip_code_prefix' # Zip codes are identifiers
]

for df_name, df in dataframes.items():
    print(f"\n--- Visualizando Outliers para DataFrame: {df_name} ---")
    numerical_cols = df.select_dtypes(include=['number']).columns

    # Filter out excluded columns
    cols_to_analyze = [col for col in numerical_cols if col not in excluded_cols_from_outliers]

    if not cols_to_analyze:
        print(f"No hay columnas numéricas relevantes para visualizar outliers en {df_name} después de excluir IDs/secuenciales.")
        continue

    for col in cols_to_analyze:
        outliers, lower_bound, upper_bound = detect_outliers_iqr(df, col)
        if not outliers.empty:
            plt.figure(figsize=(10, 6))
            sns.boxplot(y=df[col])
            plt.title(f'Box Plot de {col} en {df_name} (con Outliers)')
            plt.ylabel(col)
            plt.axhline(lower_bound, color='r', linestyle='--', label=f'Límite Inferior (Q1 - 1.5*IQR): {lower_bound:.2f}')
            plt.axhline(upper_bound, color='g', linestyle='--', label=f'Límite Superior (Q3 + 1.5*IQR): {upper_bound:.2f}')
            plt.legend()
            plt.grid(True, linestyle='--', alpha=0.6)
            plt.show()
        else:
            print(f"  No se encontraron outliers para la columna '{col}' en {df_name} según el método IQR (excluyendo IDs/secuenciales).")
```

## 6. Consolidación de Datos: Creación de `OrdersFull`

Procederemos a consolidar todos los DataFrames en uno solo, `OrdersFull`, utilizando operaciones `merge`. Esta consolidación es crucial para análisis posteriores que requieran una visión completa de los pedidos, clientes, productos y vendedores.

Para el `geolocation_df`, que contiene múltiples entradas para cada código postal, primero lo agregaremos para obtener una latitud y longitud promedio, y la ciudad/estado más frecuente por código postal. Luego, fusionaremos esta información geolocalizada por separado para los clientes y los vendedores, asegurando que el DataFrame consolidado no crezca de forma inmanejable y que la información geográfica sea representativa.

**Razonamiento**:
Para demostrar visualmente la violación de la 1FN, seleccionaré un `order_id` que tenga múltiples `order_item_id` en el DataFrame `OrdersFull` y mostraré la información a nivel de orden duplicada para cada ítem.



```python
# Aggregate geolocation data to get a single entry per zip code prefix (e.g., mean lat/lng, mode city/state)
# This prevents an explosion of rows when merging directly with many-to-many relationships.

geolocation_agg_df_customer = geolocation_df.groupby('geolocation_zip_code_prefix').agg(
    customer_geolocation_lat=('geolocation_lat', 'mean'),
    customer_geolocation_lng=('geolocation_lng', 'mean'),
    customer_geolocation_city=('geolocation_city', lambda x: x.mode()[0] if not x.mode().empty else None),
    customer_geolocation_state=('geolocation_state', lambda x: x.mode()[0] if not x.mode().empty else None)
).reset_index().rename(columns={'geolocation_zip_code_prefix': 'customer_zip_code_prefix'})

geolocation_agg_df_seller = geolocation_df.groupby('geolocation_zip_code_prefix').agg(
    seller_geolocation_lat=('geolocation_lat', 'mean'),
    seller_geolocation_lng=('geolocation_lng', 'mean'),
    seller_geolocation_city=('geolocation_city', lambda x: x.mode()[0] if not x.mode().empty else None),
    seller_geolocation_state=('geolocation_state', lambda x: x.mode()[0] if not x.mode().empty else None)
).reset_index().rename(columns={'geolocation_zip_code_prefix': 'seller_zip_code_prefix'})


# Start with orders_df as the base for the full dataset
OrdersFull = orders_df.copy()

# Merge with customers_df on customer_id
OrdersFull = pd.merge(OrdersFull, customers_df, on='customer_id', how='left')

# Merge with order_payments_df on order_id
OrdersFull = pd.merge(OrdersFull, order_payments_df, on='order_id', how='left')

# Merge with order_reviews_df on order_id
OrdersFull = pd.merge(OrdersFull, order_reviews_df, on='order_id', how='left')

# Merge with order_items_df on order_id
# Note: An order can have multiple items, so this merge will result in more rows
OrdersFull = pd.merge(OrdersFull, order_items_df, on='order_id', how='left')

# Merge with products_df on product_id
OrdersFull = pd.merge(OrdersFull, products_df, on='product_id', how='left')

# Merge with product_category_name_translation_df on product_category_name
OrdersFull = pd.merge(OrdersFull, product_category_name_translation_df, on='product_category_name', how='left')

# Merge with sellers_df on seller_id
OrdersFull = pd.merge(OrdersFull, sellers_df, on='seller_id', how='left')

# Merge customer geolocation data
OrdersFull = pd.merge(OrdersFull, geolocation_agg_df_customer, on='customer_zip_code_prefix', how='left')

# Merge seller geolocation data
OrdersFull = pd.merge(OrdersFull, geolocation_agg_df_seller, on='seller_zip_code_prefix', how='left')


print("Shape of the merged DataFrame (OrdersFull):", OrdersFull.shape)
print("\nInfo of the merged DataFrame (OrdersFull):")
OrdersFull.info()
print("\nHead of the merged DataFrame (OrdersFull):")
display(OrdersFull.head())
```

### Violaciones de 1FN en OrdersFull

La Primera Forma Normal (1FN) establece que cada columna en una tabla debe contener valores atómicos y que no debe haber grupos repetitivos de columnas. En el contexto de una base de datos relacional, esto suele significar que cada celda debe contener un único valor, y que cada fila debe representar de forma única una entidad sin duplicar información correspondiente a una sub-entidad relacionada.

El DataFrame `OrdersFull` se creó al combinar varias tablas y, de manera crítica, implicó la unión de `order_items_df` con `orders_df`. Dado que un `order_id` puede tener múltiples `order_item_id` (cada uno representando un producto o cantidad diferente dentro de la misma orden), la operación de combinación duplica efectivamente toda la información a nivel de orden (como `customer_id`, `order_status`, `order_purchase_timestamp`, `payment_type`, `payment_value`, `review_score`, etc.) por cada ítem asociado a esa orden. Esto conduce a una violación de la 1FN porque los atributos a nivel de orden no son atómicos respecto a una fila que representa un `order_item`, y conforman un grupo repetitivo para cada ítem dentro de una orden.

Por ejemplo, si una orden `X` tiene dos ítems, `A` y `B`, el DataFrame `OrdersFull` tendrá dos filas para la orden `X`, una para el ítem `A` y otra para el ítem `B`. Ambas filas contendrán información idéntica en todas las columnas relacionadas con la orden en sí (por ejemplo, `order_id`, `customer_id`, `order_status`, `payment_value`, etc.), generando redundancia.

### Violaciones de 2FN en OrdersFull

La Segunda Forma Normal (2FN) requiere que una tabla esté en 1FN y que todos los atributos no clave sean completamente dependientes funcionalmente de la *totalidad* de la clave primaria. Si una tabla tiene una clave primaria compuesta (una clave primaria conformada por dos o más columnas), ocurre una violación de 2FN cuando un atributo no clave depende únicamente de una *parte* de la clave primaria compuesta.

En nuestro DataFrame `OrdersFull`, si consideramos (`order_id`, `order_item_id`) como una posible clave primaria compuesta para el concepto de 'ítem de orden', podemos identificar varias violaciones de 2FN. Muchos atributos en `OrdersFull` son funcionalmente dependientes únicamente de `order_id`, no de la clave compuesta completa (`order_id`, `order_item_id`). Estos incluyen:

*   **Detalles del Cliente**: `customer_id`, `customer_unique_id`, `customer_zip_code_prefix`, `customer_city`, `customer_state`, y todas las columnas de `customer_geolocation` (`customer_geolocation_lat`, `customer_geolocation_lng`, `customer_geolocation_city`, `customer_geolocation_state`). Estos atributos describen al cliente que realizó la orden y dependen únicamente de `order_id`, no de `order_item_id`.

*   **Detalles de la Orden**: `order_status`, `order_purchase_timestamp`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`, `order_estimated_delivery_date`, `order_purchase_year_month`. Estas son propiedades de la orden en sí, independientemente de los ítems individuales que la componen.

*   **Detalles del Pago**: `payment_sequential`, `payment_type`, `payment_installments`, `payment_value`. Estos describen cómo se pagó la orden completa y dependen únicamente de `order_id`.

*   **Detalles de la Reseña**: `review_id`, `review_score`, `review_comment_title`, `review_comment_message`, `review_creation_date`, `review_answer_timestamp`. Estos atributos están relacionados con la reseña de la orden completa, no específicamente con un ítem particular.

Todos estos atributos a nivel de orden se están repitiendo para cada `order_item_id` asociado a un `order_id` dado. Esta redundancia es una clara violación de 2FN porque dichos atributos no dependen de la clave compuesta completa (`order_id`, `order_item_id`), sino únicamente de la parte `order_id`.

### Violaciones de 3FN en OrdersFull

La Tercera Forma Normal (3FN) requiere que una tabla esté en 2FN y que todos los atributos no clave sean no transitivamente dependientes de la clave primaria. Esto significa que un atributo no clave no debe depender de otro atributo no clave. En términos más sencillos, la 3FN busca eliminar las dependencias transitivas.

En nuestro DataFrame `OrdersFull`, incluso después de considerar las potenciales violaciones de 2FN, aún existen atributos que exhiben dependencias transitivas. Estas generalmente ocurren cuando un atributo depende de otro atributo no clave, que a su vez depende de la clave primaria.

Las principales violaciones de 3FN en `OrdersFull` probablemente surgen de lo siguiente:

*   **Traducción del Nombre de la Categoría del Producto**: `product_category_name_english` depende de `product_category_name`, que es un atributo no clave (asumiendo que `product_id` es la clave primaria para los detalles del producto, o (`order_id`, `order_item_id`) para el concepto de ítem de orden). `product_category_name` depende a su vez de `product_id`. Esto crea una dependencia transitiva: `(product_id) -> product_category_name -> product_category_name_english`. La traducción al inglés no depende directamente de `product_id` (ni de la clave compuesta).

*   **Datos de Geolocalización**: Los datos de geolocalización agregados (`customer_geolocation_lat`, `customer_geolocation_lng`, `customer_geolocation_city`, `customer_geolocation_state`) tanto para clientes como para vendedores dependen de `customer_zip_code_prefix` y `seller_zip_code_prefix` respectivamente. Estos códigos postales son atributos no clave (por ejemplo, `customer_zip_code_prefix` depende de `customer_id`, el cual depende de `order_id`). Esto conforma una dependencia transitiva: `(customer_id) -> customer_zip_code_prefix -> atributos_de_geolocalizacion_del_cliente`. Lo mismo aplica para la geolocalización del vendedor.

*   **ID Único del Cliente**: `customer_unique_id` depende de `customer_id`, que es un atributo no clave cuando se considera el DataFrame `OrdersFull`, donde `order_id` o (`order_id`, `order_item_id`) harían parte de la clave primaria. Si `customer_id` es la clave primaria para la información del cliente, entonces `customer_unique_id` (que representa a una persona única a través de las compras) dependería de él.

Estas dependencias transitivas introducen redundancia y potenciales anomalías de actualización. Por ejemplo, si la traducción al inglés de un `product_category_name` cambia, tendría que ser actualizada en múltiples filas dentro de `OrdersFull`.

### Esquema Relacional Normalizado Propuesto

Para abordar las violaciones de 1FN, 2FN y 3FN identificadas, un esquema relacional normalizado implicaría descomponer el DataFrame `OrdersFull` en varias tablas distintas, cada una representando una única entidad y cumpliendo con la 3FN. Este enfoque busca eliminar la redundancia de datos y mejorar la integridad de los mismos.

**1. Entidades Normalizadas Principales (3FN):**

*   **Tabla de Clientes (`customers_df_normalized`):**

    *   Atributos: `customer_unique_id`, `customer_zip_code_prefix`, `customer_city`, `customer_state`
    *   Justificación: Esta tabla almacenaría toda la información específica del cliente, eliminando la duplicación de `customer_unique_id`, `customer_zip_code_prefix`, `customer_city` y `customer_state` en la tabla `orders`. `customer_id` es la clave primaria natural.

*   **Tabla de Órdenes (`orders_df_normalized`):**
    *   Clave Primaria: `order_id`
    *   Clave Foránea: `customer_id` (referencia a `customers_df_normalized`)
    *   Atributos: `customer_id`, `order_status`, `order_purchase_timestamp`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`, `order_estimated_delivery_date`
    *   Justificación: Esta tabla contendría todos los detalles a nivel de orden. Todos los atributos no clave son completamente dependientes de `order_id`, resolviendo las violaciones de 2FN presentes en `OrdersFull`, donde los detalles a nivel de orden se repetían para cada ítem de la orden.

*   **Tabla de Ítems de Orden (`order_items_df_normalized`):**
    *   Clave Primaria Compuesta: `(order_id, order_item_id)`
    *   Claves Foráneas: `order_id` (referencia a `orders_df_normalized`), `product_id` (referencia a `products_df_normalized`), `seller_id` (referencia a `sellers_df_normalized`)
    *   Atributos: `order_id`, `order_item_id`, `product_id`, `seller_id`, `shipping_limit_date`, `price`, `freight_value`
    *   Justificación: Esta tabla captura la relación de muchos a muchos entre órdenes y productos. Aborda directamente la violación de 1FN al garantizar que cada fila represente un `order_item` único, y que los detalles a nivel de orden no se repitan aquí.

*   **Tabla de Productos (`products_df_normalized`):**
    *   Clave Primaria: `product_id`
    *   Clave Foránea: `product_category_name` (referencia a `product_category_translation_df_normalized`)

    *   Justificación: Esta tabla contiene la información única del producto, y `product_category_name` enlaza con la traducción de la categoría.

*   **Tabla de Traducción de Categorías de Producto (`product_category_translation_df_normalized`):**
    *   Clave Primaria: `product_category_name`
    *   Atributos: `product_category_name_english`
    *   Justificación: Resuelve la violación de 3FN relacionada con la traducción de la categoría. `product_category_name_english` es transitivamente dependiente de `product_category_name`.

*   **Tabla de Vendedores (`sellers_df_normalized`):**
    *   Clave Primaria: `seller_id`
    *   Atributos: `seller_zip_code_prefix`, `seller_city`, `seller_state`
    *   Justificación: Almacena la información única del vendedor, de forma similar a la tabla `Customers`.

*   **Tabla de Pagos de Orden (`order_payments_df_normalized`):**
    *   Clave Primaria Compuesta: `(order_id, payment_sequential)`
    *   Clave Foránea: `order_id` (referencia a `orders_df_normalized`)
    *   Atributos: `payment_type`, `payment_installments`, `payment_value`
    *   Justificación: Captura los detalles del pago, los cuales pueden ser múltiples para una sola orden (de ahí el `payment_sequential`). Todos los atributos dependen de la clave compuesta.

*   **Tabla de Reseñas de Orden (`order_reviews_df_normalized`):**

**2. Escenarios de Desnormalización Estratégica y Justificaciones:**

Si bien un esquema completamente normalizado es ideal para sistemas transaccionales (OLTP), las consultas analíticas (OLAP) a menudo se benefician de cierto nivel de desnormalización para reducir uniones (joins) y mejorar el rendimiento de las consultas. A continuación, se presentan instancias específicas donde la desnormalización podría ser beneficiosa:

*   **Incrustar el Nombre en Inglés de la Categoría de Producto en la Tabla `Products` (o en `Order Items` para reportería):**
    *   Escenario: Si los reportes analíticos requieren con frecuencia mostrar los nombres de las categorías de productos en inglés junto con los detalles del producto.
    *   Justificación: En lugar de unir `products_df_normalized` con `product_category_translation_df_normalized` de manera repetida, se podría añadir una columna `product_category_name_english` directamente a `products_df_normalized` o a una tabla analítica derivada. Esto evita una operación de unión, acelerando las consultas a costa de una leve redundancia y una potencial anomalía si una traducción cambia (aunque esto es poco probable para nombres de categorías).

*   **Incluir `customer_city`, `customer_state`, `seller_city`, `seller_state` directamente en las Tablas Analíticas de `Orders` u `Order Items`:**
    *   Escenario: Para análisis geográficos comunes, como ventas por estado del cliente o ubicación del vendedor, sin necesidad de unir las tablas `Customers`, `Sellers` y `Geolocation` cada vez.
    *   Justificación: Esto evita múltiples uniones, mejorando el rendimiento de las consultas para reportería agregada basada en ubicación. La compensación es la redundancia, ya que esta información se encuentra disponible a través de los códigos postales enlazados a la tabla `Geolocation`. Esto es aceptable para vistas analíticas en las que las anomalías de actualización de datos son menos críticas que la velocidad de consulta.

*   **Crear una Tabla de Hechos Analítica (por ejemplo, `OrdersFact`):**
    *   Escenario: Para consultas analíticas complejas que involucran rendimiento de ventas, comportamiento del cliente y desempeño del vendedor, las cuales requieren combinar datos de muchas tablas normalizadas distintas.
    *   Justificación: El DataFrame `OrdersFull` original, a pesar de sus violaciones de normalización, sirve como un excelente ejemplo de una tabla desnormalizada apta para uso analítico directo (por ejemplo, en un esquema en estrella de un data warehouse, donde `OrdersFull` podría ser una tabla de hechos). Pre-une todas las dimensiones necesarias (cliente, producto, vendedor, tiempo, etc.), lo cual la hace muy rápida para reportería y dashboards. Esto reduce significativamente la complejidad y el tiempo de ejecución de las consultas para patrones analíticos comunes, sacrificando el rendimiento de escritura y la consistencia de actualización a favor de la optimización de lectura.

**3. Enfoque Estratégico General:**

El enfoque estratégico implica un modelo híbrido que aprovecha tanto la normalización como la desnormalización, dependiendo del caso de uso:

1.  **Mantener un Núcleo Normalizado para el Procesamiento Transaccional y la Integridad de los Datos:** La base de datos operativa subyacente (OLTP) debería ceñirse principalmente a un esquema normalizado en 3FN. Esto asegura el más alto nivel de integridad de los datos, reduce la redundancia, minimiza el espacio de almacenamiento para datos que se actualizan con frecuencia y simplifica el mantenimiento de los datos. Esto es crucial para aplicaciones que involucran agregar, actualizar o eliminar registros (por ejemplo, procesamiento de órdenes, gestión de clientes).

2.  **Emplear Desnormalización Estratégica para Reportería Analítica (OLAP):** Para propósitos analíticos, se puede crear un almacén de datos separado (por ejemplo, un data warehouse o un data mart). Aquí es donde la desnormalización estratégica se vuelve altamente valiosa. El DataFrame `OrdersFull`, o una versión ligeramente refinada del mismo, puede servir como una vista analítica principal o tabla de hechos dentro de este entorno OLAP. Esto permite que los analistas y las herramientas de inteligencia de negocios consulten conjuntos de datos pre-unidos e integrales de manera rápida, sin afectar el rendimiento ni la integridad del sistema transaccional.

**Justificación del Enfoque Híbrido:**

*   **Integridad de Datos vs. Rendimiento de Consultas:** La normalización prioriza la integridad de los datos, asegurando que estos se almacenen de manera lógica y consistente. La desnormalización prioriza el rendimiento de las consultas al reducir la necesidad de uniones complejas. Un enfoque híbrido nos permite lograr ambos objetivos: integridad en el sistema fuente y rendimiento en el sistema de reportería.
*   **Separación de Responsabilidades:** Al contar con una capa transaccional normalizada y una capa analítica desnormalizada, separamos las responsabilidades. Los cambios en el sistema transaccional no impactan directamente los reportes analíticos, y la optimización de las consultas analíticas no compromete los datos operativos.
*   **Escalabilidad:** Las bases de datos OLTP están optimizadas para operaciones de escritura, mientras que las bases de datos OLAP están optimizadas para operaciones de lectura. Un modelo híbrido permite que cada sistema escale de forma independiente y sea optimizado para su carga de trabajo principal.

En conclusión, el DataFrame `OrdersFull` inicial, a pesar de sus violaciones de 1FN, 2FN y 3FN, ofrece un sólido punto de partida para una tabla de hechos analítica. Si bien no es adecuado como modelo de datos transaccional principal, demuestra de manera efectiva los beneficios de la desnormalización para casos de uso específicos, como la inteligencia de negocios integral y la reportería analítica, donde la recuperación rápida de datos a través de múltiples dimensiones es primordial. La estrategia recomendada es mantener los DataFrames originales cargados sin procesar (como `customers_df`, `orders_df`, etc.) como la capa 'fuente' o de 'staging' normalizada, y luego utilizar `OrdersFull` (o tablas desnormalizadas derivadas y ligeramente más optimizadas) para necesidades analíticas complejas.

### Discutiendo los Trade-offs de la Normalización y la Desnormalización Estratégica

#### Ventajas de una Normalización Alta
Una normalización alta, que se adhiere a principios como la 3FN, ofrece beneficios significativos para nuestro conjunto de datos de e-commerce:

*   **Integridad de los Datos**: Al descomponer el DataFrame `OrdersFull` en tablas separadas y bien definidas (por ejemplo, `orders`, `customers`, `order_items`, `products`, `sellers`, `payments`, `reviews`, `geolocation`, `product_category_translation`), nos aseguramos de que cada pieza de información se almacene en un único lugar. Esto previene anomalías de actualización; por ejemplo, si la dirección de un cliente cambia, solo necesita actualizarse en la tabla `customers`, no en cada orden que haya realizado. De manera similar, los detalles del producto (como las dimensiones) se almacenan una sola vez en la tabla `products`.
*   **Redundancia Reducida**: El DataFrame `OrdersFull` demuestra claramente la redundancia. Todos los detalles a nivel de orden (`customer_id`, `order_status`, `order_purchase_timestamp`, `payment_type`, `payment_value`, `review_score`) se duplican por cada ítem individual (`order_item_id`) dentro de una orden. De forma similar, los detalles del producto (`product_category_name`, `product_name_lenght`, `product_weight_g`) se repiten por cada `order_item` que referencia el mismo `product_id`. La normalización elimina esta extensa duplicación, conduciendo a un almacenamiento más eficiente y a menos errores.
*   **Mantenimiento más Sencillo**: Con los datos almacenados de manera lógica y sin redundancia, la base de datos es más fácil de mantener. Las inserciones, eliminaciones y actualizaciones se simplifican. Por ejemplo, agregar un nuevo producto no requiere modificar las tablas `order_items` u `orders` existentes.
*   **Menos Anomalías de Actualización**: Los cambios a un atributo específico (por ejemplo, el precio de un producto, o la ciudad de un vendedor) solo necesitan realizarse en un registro dentro de su respectiva tabla normalizada. Esto previene inconsistencias que podrían surgir si las copias redundantes de los datos no se actualizan de manera uniforme a través de una estructura desnormalizada como `OrdersFull`.

#### Desventajas de una Normalización Alta
Aunque es beneficiosa, una normalización excesiva también conlleva inconvenientes:

*   **Mayor Complejidad en las Uniones**: Para reconstruir una vista completa de una orden a partir de un esquema altamente normalizado (por ejemplo, detalles de la orden, información del cliente, todos los ítems comprados, detalles del producto para cada ítem, información del vendedor, pago y detalles de reseñas), deben unirse muchas tablas. Esto es precisamente lo que implica la creación de `OrdersFull`, involucrando numerosas operaciones de `pd.merge`. Para consultas analíticas complejas o tareas operativas que requieren una vista integral, la cantidad de uniones (joins) puede volverse sustancial.
*   **Potencial Sobrecarga de Rendimiento**: Más uniones se traducen en una mayor sobrecarga computacional y, potencialmente, tiempos de ejecución de consultas más lentos. Cada operación de unión requiere que el sistema de base de datos realice trabajo adicional, lo cual puede convertirse en un cuello de botella para reportería, data warehousing o aplicaciones altamente interactivas que necesitan mostrar datos compuestos de manera rápida. El DataFrame `OrdersFull`, a pesar de su violación de 1FN, está diseñado precisamente para este propósito: proporcionar una tabla única lista para usar en el análisis, que de otra forma requeriría múltiples uniones.

#### Escenarios de Desnormalización Estratégica para este Conjunto de Datos de E-commerce
Dados los trade-offs, la desnormalización estratégica puede ser beneficiosa para casos de uso específicos en este conjunto de datos de e-commerce. El objetivo es mejorar el rendimiento de las consultas para datos combinados de acceso frecuente o reportes críticos en términos de rendimiento, minimizando los riesgos de problemas de integridad de datos.

1.  **Detalles del Producto con Ítems de Orden**: Al analizar tendencias de ventas, popularidad de productos o ingresos por categoría de producto, las consultas frecuentemente unen `order_items_df` con `products_df` y `product_category_name_translation_df`. Por ejemplo, para recuperar rápidamente el nombre en inglés de la categoría del producto junto con cada ítem de la orden:
    *   **Estrategia de Desnormalización**: Incluir `product_category_name` y `product_category_name_english` directamente dentro de la tabla `order_items` o en una tabla `products` que esté pre-unida con `product_category_name_translation_df`.
    *   **Justificación**: Esto eliminaría una unión hacia `product_category_name_translation_df` para casi todos los análisis relacionados con productos, acelerando consultas para reportes como "Productos más vendidos por categoría en inglés" o "Precio promedio por categoría."

2.  **Datos de Ubicación de Cliente/Vendedor (Ciudad/Estado)**: Tanto `customers_df` como `sellers_df` ya contienen columnas `_city` y `_state`. El DataFrame `OrdersFull` también incluye datos agregados de geolocalización `lat/lng` y `city/state` provenientes de `geolocation_agg_df_customer` y `geolocation_agg_df_seller` con base en los códigos postales.
    *   **Estrategia de Desnormalización**: Para la mayoría de los propósitos analíticos, tener las columnas `customer_city`, `customer_state`, `seller_city` y `seller_state` directamente en las tablas `customers` y `sellers` (como están actualmente) constituye una forma de desnormalización si se considera que la fuente autoritativa para ciudad/estado es `geolocation_df`. `OrdersFull` ya las incorpora para acceso inmediato.
    *   **Justificación**: Recuperar la ciudad y el estado para clientes y vendedores es un requerimiento muy común. Incrustar directamente estos atributos altamente estables en las tablas `customers` y `sellers` evita una unión hacia los datos de `geolocation` por cada consulta, mejorando significativamente el rendimiento para análisis basados en ubicación y aplicaciones de mapeo.

3.  **Resumen de Órdenes para Dashboards/Reportería**: Muchos dashboards de inteligencia de negocios requieren una vista integral de las órdenes, incluyendo datos demográficos del cliente, tipo de pago, valor total y estado de entrega.
    *   **Estrategia de Desnormalización**: Crear una "Tabla de Hechos" dedicada (similar a `OrdersFull` pero potencialmente más optimizada para necesidades específicas de reportería, quizás excluyendo campos de texto altamente variables como `review_comment_message`) que pre-una `orders`, `customers`, `order_payments` y `order_reviews`. Esta tabla probablemente haría parte de un esquema en estrella de un data warehouse.
    *   **Justificación**: `OrdersFull` por sí mismo cumple bien con este propósito, habilitando consultas analíticas complejas a través de diversas dimensiones sin tener que realizar docenas de uniones sobre la marcha. Esto acelera dramáticamente las consultas agregadas para reportes como "Ventas mensuales por estado del cliente y tipo de pago" o "Puntaje promedio de reseñas por categoría de producto."

En esencia, mientras que un esquema altamente normalizado garantiza la integridad de los datos y reduce la redundancia a nivel transaccional, una vista desnormalizada cuidadosamente considerada (como `OrdersFull` o tablas de hechos especializadas) es crucial para el rendimiento analítico y la simplificación de consultas en un contexto de análisis de datos.
